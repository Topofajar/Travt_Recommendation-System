# Create data rating

```python
import numpy as np
import pandas as pd
```

```python
# Read dataset
data = pd.read_csv("https://raw.githubusercontent.com/Topofajar/Travt_Recommendation-System/main/Dataset/Tourist%20and%20Resto%20Data.csv")
data.head(5)
```

```python
# Load data review
data_coba = data.copy()
kolom = ['place_id', 'rating_per_score_1', 'rating_per_score_2', 'rating_per_score_3', 'rating_per_score_4', 'rating_per_score_5']
data_baru = data_coba[kolom]
data_baru.head(5)
```

```python
data_baru.drop(columns=['place_id'], inplace=True)
```

```python
# Make random rating dataset based on data review
import random

user_id = []
place_id = []
rating = []


for i in range(len(data_baru)):
  sum_row = data_baru.iloc[i].sum()
  user_id.extend(random.sample(range(1, 27500), k=sum_row))
  place_id.extend([i + 1] * sum_row)
  for j, k in zip(data_baru.iloc[i], range(1,6)):
    rating.extend([k] * j)

# user_id = [elem for sublist in user_id for elem in sublist]
# place_id = [elem for sublist in place_id for elem in sublist]
# rating = [elem for sublist in rating for elem in sublist]

dataset_rating = pd.DataFrame({
    'user_id': user_id,
    'place_id': place_id,
    'rating': rating
})

dataset_rating.shape
```

# Foto

```python
import pandas as pd
foto = pd.read_csv('/content/Cleaning data 1 - Coretan Foto (6).csv')
foto
```

```python
id = []
url = []
for i in foto['query'].unique():
  id.append(i)
  url.append(foto.loc[foto['query'] == str(i), 'photo_url_big'].tolist())

data_foto = pd.DataFrame({
    'id': id,
    'url': url
})

data_foto.head(5)
```

```python
data_foto['url']
```

```python
data_foto['url'][1]
```

```python

# Menentukan panjang maksimum dari sub-list
max_length = max(len(sublist) for sublist in data_foto['url'])

# Menambahkan nilai None untuk melengkapi sub-list yang lebih pendek
for sublist in data_foto['url']:
    sublist.extend([None] * (max_length - len(sublist)))

# Membuat kolom tambahan berdasarkan elemen dalam sub-list
url_df = pd.DataFrame(data_foto['url'].tolist(), columns=[f'Kolom{i+1}' for i in range(max_length)])

# Menggabungkan DataFrame asli dengan DataFrame baru
result_df = pd.concat([data_foto.drop(columns=['url']), url_df], axis=1)

# Menampilkan DataFrame hasil
result_df.head(5)
```

```python
result_df.to_excel('data_foto.xlsx', index=False)
```
