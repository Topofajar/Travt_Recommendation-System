# Buat rekomendasi untuk user lama

```python
num_items = 742
def recommend_items(model, user_id, num_recommendations=10):
    user_input = np.array([user_id] * num_items)

    place_id_input = np.array(data_encode['place_id'])-1
    category_input = np.array(data_encode['category'])
    city_input = np.array(data_encode['city'])
    rating_input = np.array(data_encode['rating'])
    verified_input = np.array(data_encode['verified'])

    predictions = model.predict([user_input, place_id_input, category_input, city_input, rating_input, verified_input])
    predicted_ratings = pd.Series(predictions.flatten(), index=place_id_input)

    top_items = predicted_ratings.nlargest(num_recommendations).index
    index_place = data_encode['place_id'].astype('category').cat.categories[top_items]
    filtered_data = data_place.loc[data_place['place_id'].isin(index_place), ['name', 'city']]
    return index_place, filtered_data
```

```python
index, recommendations = recommend_items(model, user_id=0)
print("Recommendations for User 1 : ", index, "\n", recommendations)
```

# Buat rekomendasi untuk user baru

```python
def rekomendasi(model_try, rating_input, place_input, category_input, city_input, rating_place_input, verified_input):
  user_input = [27498]*len(place_input)

  # Fit the model with the input data
  model_try.fit([np.array(user_input), np.array(place_input), np.array(category_input), np.array(city_input),
            np.array(rating_place_input), np.array(verified_input)],
            np.array(rating_input), epochs=1, verbose=0)

  user_input_full = np.array([27498] * len(np.arange(1, 743)))

  # Predict ratings for all places in batch
  pred_ratings = model_try.predict([user_input_full, np.array(data_encode['place_id'])-1, np.array(data_encode['category']),
                                    np.array(data_encode['city']), np.array(data_encode['rating']), np.array(data_encode['verified'])])

  # Get top n recommendations
  rekomendasi = sorted(zip(np.arange(1, 743), pred_ratings.flatten()), key=lambda x: x[1], reverse=True)[:10]
  rekomendasi_place_ids = [place for place, rating in rekomendasi]
  return rekomendasi_place_ids
```

```python
# Sample input data
id_ = 'cadfasdfadgfssss'
place_input = [1]
rating_input = [5]
category_input = [0]
city_input = [0]
rating_place_input = [0]
verified_input = [0]
model1 = model

rekomendasi(model1, place_input, rating_input, category_input, city_input, rating_place_input, verified_input)
```
